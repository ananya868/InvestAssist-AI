```python
"""Workflow"""

# INPUT-DATA
# Investor Profile Data
# Investor Risk Data 
# Investor Portfolio Data  


### Make sure to write reusable code
### So that we can modify the agent when needed 
### Try and test out experiments for best results

# Agent
# 1. Market data Specifier Agent
#   -> Stocks, Bonds, Mutual funds, shares, debts,
#  FDs, RDs, PPFs, Real state, Commodities, Gold, Crypto, 
#  Exchange traded funds, Derivatives, Pension Plans

# 2. Data Collector Agent 
#   -> Collects the data for above listed items and provide a complete information on the above 
#   -> Makes each data separated
#   -> Provides the data in a structured format

# 3. Data Cleaner and Organizer 
#   -> Writes the data in a professional insightful manner for making it usable 
#   -> Makes it savable for a pdf file (documents)
```

```python
"""Data Specifier Agent"""
# Specifies what data the next agent has to fetch from the internet
# 

data_specifier_agent = Agent(
    role = 'Data Specifier',
    goal = 'You compile a comprehensive list of asset/investment options for investor portfolio optimization.',
    backstory = 'Act as a Data Specifier Agent for company specializing in investors portfolio optimization.'
        'Your primary function is to compile a comprehensive and clear list of various asset/investment options for use by the next agent in the process.',
    allow_delegation = True, 
    verbose = 0,
)

data_specifier_task = Task(
    description = ' You will conduct in-depth research on various asset classes in the list: {investment_list}. gather and organize relevant data on these investment options.'
                  ' Your expertise in financial analysis and market trends will ensure that our clients have access to the most suitable investment opportunities,'
                  ' tailored to their specific needs.',
    expected_output = 'A list of investment options', 
    agent = data_specifier_agent, 
)



"""Data Collector Agent"""
# Collects data from the internet based on instruction provided by the previous agent
#

data_collector_agent = Agent(
    role = 'Data Collector',
    goal = 'Data Collector, specializing in gathering real-time data on various assets and investment options from the internet for the purpose of contributing to a portfolio optimization system.',
    backstory = 'As a Data researcher, you will be responsible for gathering and compiling real-time data on a variety of assets and investment options given to you. '
                'Your primary function will be to provide up-to-date and comprehensive information that can be used to optimize investment portfolios.'
                'You will need to conduct thorough research, analyze market trends, and present your findings in a clear and concise manner.'
                'This data will serve as a valuable resource for the next agent in the process, who will use it to make informed investment decisions',
    allow_delegation = True,
    verbose = 1,
)

from crewai_tools import EXASearchTool

# Initialize the tool for internet searching capabilities
exa_tool = EXASearchTool()

data_collector_task = Task(
    description = 'Your primary function is to search the internet and compile current data on a list of assets and investment options provided to you.'
                    'This data must be from the current date: {current_date} only and should be detailed enough to offer valuable insights for portfolio optimization.'
                    'The output of your work will serve as a document or insight for the next agent in the process.'
                    'Focus on gathering real-time, current date data on the provided list of assets and investment options.'
                    'Ensure the information is comprehensive, covering key aspects that would influence investment decisions such as market performance, trends, and forecasts.'
                    'Present the data in a structured, easy-to-understand format that can be directly used for portfolio optimization.',
    expected_output = 'A document or insight containing current data on the provided list of assets and investment options'
                    'This insight document include real-time insights and analysis on the list of assets and investments'
                    'This document should be formatted in such a way that there is separate paras for each investment options mentioned before.',
    tools = [exa_tool],
    agent = data_collector_agent,
)



"""Data Formatter Agent"""
# Formats the data collected by the previous agent into a professional document
#

data_formatter_agent = Agent(
    role = 'Data Formatter',
    goal = 'As a data formatter/writer, you take in a text corpus(multiple paras) and format the data according to the instructions provided',
    backstory = 'You are a specialist in writing and you write raw information into specified format'
                'You role is to follow the instructions of format and write as required',
    allow_delegation = False,
    verbose = 1,
)

data_formatter_task = Task(
    description = 'Your primary function is to take the raw text data provided to you and format it into a professional document.'
                    'You will need to structure the information in a clear and concise manner, ensuring that it is easy to read and understand.'
                    'The document should be well-organized, with each section clearly labeled and formatted according to these guidelines: {guidelines}.'
                    'The final document should be polished and ready for presentation to clients or stakeholders.',
    expected_output = 'A well-formatted document containing the data provided to you in a professional and '
                    'insightful manner and it should strictly follow these data format guidelines: {guidelines}.',
    agent = data_formatter_agent,
)
```

```python
crew = Crew(
    agents = [data_specifier_agent, data_collector_agent, data_formatter_agent],
    
    tasks=[data_specifier_task, data_collector_task, data_formatter_task],
	
    verbose=2,
	memory=True
)
```

```python
inputs = {
    'investment_list': 'Stocks, Bonds, Mutual funds, shares, debts, FDs, RDs, PPFs, Real state, Commodities, Gold, Crypto, Exchange traded funds, Derivatives, Pension Plans',
    'current_date': '2022-10-10',
    'guidelines': 'Markdown format. Make the title of investment name as subheading and the content as normal text. This markdown should contain separate sections/paras for each investment options that was provided. Dont make it too big. Maximum word limit is 600'
}

result = crew.kickoff(inputs=inputs)
```
